```python
import json
import pandas as pd
import matplotlib.pyplot as plt
import requests

def obtener_datos_quiz(api_url):
    try:
        response = requests.get(api_url)
        response.raise_for_status()
        data = response.json()
        print(json.dumps(data, indent=4))  # Imprimir la respuesta JSON
        return data
    except requests.exceptions.RequestException as e:
        print(f"Error al obtener datos de la API: {e}")
        return None

api_url = 'http://localhost:3000/preguntes'
dades = obtener_datos_quiz(api_url)

if dades:
    # Comprobando si las claves existen en los datos
    if 'preguntes' in dades and isinstance(dades['preguntes'], list):
        for pregunta in dades['preguntes']:
            if 'encerts' not in pregunta or 'intents' not in pregunta:
                print(f"Faltan las claves en la pregunta ID {pregunta['id']}")
        
        # Crear el DataFrame solo si todas las preguntas tienen las claves
        df = pd.DataFrame(dades['preguntes'])

        # Calcular el porcentaje de aciertos si las columnas existen
        if 'encerts' in df.columns and 'intents' in df.columns:
            df['percentatge_encerts'] = (df['encerts'] / df['intents']) * 100
            
            # Crear gráficos de barras
            plt.figure(figsize=(10, 6))
            plt.bar(df['pregunta'], df['percentatge_encerts'], color='skyblue')
            plt.xlabel('Preguntas')
            plt.ylabel('Porcentaje de aciertos (%)')
            plt.title('Porcentaje de aciertos por pregunta')
            plt.xticks(rotation=45)
            plt.tight_layout()
            plt.show()
        else:
            print("Las columnas 'encerts' e 'intents' no están disponibles en el DataFrame.")
    else:
        print("No se encontraron las preguntas en los datos.")
else:
    print("No se pudieron obtener los datos del cuestionario.")
```
